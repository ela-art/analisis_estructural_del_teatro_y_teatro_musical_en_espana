```python
import pandas as pd          # Manejo y análisis de datos en tablas (DataFrames)
from IPython.display import display  # Muestra objetos de forma clara en notebooks
import matplotlib.pyplot as plt      # Crear gráficos y visualizaciones
import seaborn as sns        # Gráficos estadísticos atractivos (basado en matplotlib)
from matplotlib.lines import Line2D  # Crear o personalizar líneas en gráficos
```

```python
df = pd.read_csv("porcentaje_PIB_cultura_2023.csv")
```

```python
# EDA – Porcentaje PIB Cultura 2023

# 1. Describe general
df.describe(include='all')

# 2. Comprobación básica de consistencia
print("Valores únicos en 'Administracion':")
print(df["Administracion"].unique())

print("\nValores negativos o fuera de rango:")
print(df[df["Porcentaje_PIB_2023"] < 0])

# 3. KPI: aporte total al PIB cultural
PIB_total = df["Porcentaje_PIB_2023"].sum()
print("\nAporte total al PIB cultural 2023:", PIB_total)

# 4. Visualización opcional
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.barh(df["Administracion"], df["Porcentaje_PIB_2023"])
plt.xlabel("Porcentaje del PIB dedicado a cultura (2023)")
plt.title("Aporte al PIB cultural por Administración (2023)")
plt.tight_layout()
plt.show()
```

```python
# Guardar dataset limpio y validado
df.to_csv("porcentaje_PIB_cultura_2023_limpio.csv", index=False)

print("porcentaje_PIB_cultura_2023_limpio.csv")
```
